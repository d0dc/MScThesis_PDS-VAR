### 2. Double-post-selection (PDS)
$$ y_{BTC,t} = c_{BTC} + a_{BTC,1}y_{BTC,t-1} + a_{BTC,2}y_{ETH,t-1} + u_{BTC,t} $$
$$ y_{ETH,t} = c_{ETH} + a_{ETH,1}y_{BTC,t-1} + a_{ETH,2}y_{ETH,t-1} + u_{ETH,t} $$

Since I have 2 series in the VAR system, I have a total of 5 PDS steps (4 model selections, 1 for final OLS est. of structural model) (2series x 2steps + 1VARestimation). For simplicity, I assume perfect model selection of the GDP above.

PDS general steps:
>1) Regress variable of interest on controls (reduced-form model, for prediction).
>
>2) Regress dependent variable on controls (reduced-form model, for prediction).
>
>3) Regress dependent variable on union of controls selected in two previous steps (structural-form model, for inference).

Specifically, in my case, first do PDS for BTC:
>1) Regress: $y_{ETH,t-1}=c+a_{1}y_{BTC,t-1}+u_{t}$ (i.e. $y_{ETH,t}=c+a_{1}y_{BTC,t}+u_{t}$)
>
>2) Regress: $y_{BTC,t}=c+b_{1}y_{BTC,t-1}+b_{2}y_{ETH,t-1}+v_{t}$
>
>3) Keep selected regressors in both steps and include them in VAR.

Then do PDS for ETH:
>1) Regress: $y_{BTC,t-1}=c+a_{1}y_{ETH,t-1}+u_{t}$ (i.e. $y_{BTC,t}=c+a_{1}y_{ETH,t}+u_{t}$)
>
>2) Regress: $y_{ETH,t}=c+b_{1}y_{BTC,t-1}+b_{2}y_{ETH,t-1}+v_{t}$
>
>3) Keep selected regressors in both steps and include them in VAR.

Using PDS, I've selected the relevant regressors to model BTC and ETH. I must now estimate the following VAR via OLS:

$$ y_{t} = c + Ay_{t-1} + u $$

$$ \begin{bmatrix}
y_{BTC,t} \\ y_{ETH,t}
\end{bmatrix}
=
\begin{bmatrix}
c_{BTC} \\ c_{ETH}
\end{bmatrix}
+
\begin{bmatrix}
a_{BTC,1} & 0 \\
a_{ETH,1} & a_{ETH,2}
\end{bmatrix}
\begin{bmatrix}
y_{BTC,t-1} \\ y_{ETH,t-1}
\end{bmatrix}
+
\begin{bmatrix}
u_{BTC,t} \\ u_{ETH,t}
\end{bmatrix} $$
